### Для какого вида объекта общественного питания характерно сетевое распространение? <a id='2.3'></a>

Построим сводную таблицу распределения типов объектов по сетевому признаку, чтобы выяснить, для какого вида объекта общественного питания характерно сетевое распространение:

```python
types_chain = rest_data.groupby(['object_type', 'chain']).agg({'id':'count'})
#добавим столбец с процентным соотношением
types_chain['pct'] = types_chain.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum())).round(1)
#сбросим индексы
types_chain = types_chain.reset_index()
types_chain
```

```python
#столбчатая диграмма по датафрейму types_chain
fig = px.bar(types_chain,
             x = 'object_type', y = 'id', text = 'pct', color = 'chain', barmode = 'group',
             title = 'Сетевые заведения', labels={
                     'id': "Кол-во",
                     'object_type': '',
                 'chain': 'сеть'
                 })

#изменим порядок сортировки
fig.update_layout(xaxis = {'categoryorder':'min ascending'})
#отобразим диграмму
fig.show('png');
```

Ожидаемо больше всего сетевых заведений среди фастфуда. Сетевых столовых единицы, хотя зачастую попадаются сетевые столовые в бизнес-центрах. Может быть, после этого они перестают называться столовыми?

### Выделить в отдельный столбец информацию об улице из столбца 'address' <a id='2.6'></a>

Добавим новый признак 'street' с названием улицы в датафрейм rest_data. Для этого воспользуемся регулярными выражениями:

```python
#создадим словарь сущностей для возможных типов улиц
essense = ['улица','ул','переулок','шоссе','проспект','площадь','проезд',
         'село','аллея','бульвар','набережная','тупик','линия', 'деревня']
#паттерн регулярного выражения для вычленения названия улицы из адреса
ptrn = r',\s*\b([^,]*?(?:{})\b[^,]*)[,$]+'.format('|'.join(essense))
#воспользуемся методом extract без учета регистра
rest_data['street'] = rest_data['address'].str.extract(ptrn, flags = re.I)
#проверим результат
rest_data
```

### Найти число улиц с одним объектом общественного питания. Воспользуйтесь внешней информацией и ответьте на вопрос — в каких районах Москвы находятся эти улицы? <a id='2.8'><a>

Теперь найдем улицы с лишь одним объектом общественного питания:

```python
#сводная таблица улица/кол-во id объектов методом groupby
places_num = rest_data.groupby('street').agg({'id':'count'}).sort_values(by = 'id')
#переименуем стобец 'id' в 'count'
places_num.rename(columns = {'id':'count'}, inplace = True)
#выборка с улицами для которых count = 1
only_one_place = places_num[places_num['count'] == 1].reset_index()
only_one_place
```

Получили список с 565 улицами. Видимо много коротких и расположенных в спальных районах и за МКАД улиц.

Теперь узнаем, в каких районах расположены эти улицы, воспользуемся API dadata:

```python
ones_dists = only_one_place['street']
ones = rest_data.query('street in @ones_dists')
ones['district'] = ones['address'].apply(lambda x: dadata
                                         .suggest('address', x)[0]['data']['city_district_with_type'])
ones
```

```python
ones_clean = ones.dropna().reset_index()
ones_clean
```

```python
districts_ones = (ones_clean.groupby('street').agg({'id': 'count', 'district':', '.join})
                  .sort_values(by = 'id', ascending = False)).reset_index()
districts_ones['district'] = districts_ones['district'].apply(lambda x: x.split(', ')).apply(lambda x: set(x))
districts_ones.head(10)
```

Подвердилось предположение о том, что улицы с одним заведением - достаточно короткие улицы, все улицы из первой двадцатки занимают не более одного района.

Проверим, в каких районах больше всего улиц с 1 заведением:

```python
districts_ones['district'] = districts_ones['district'].apply(lambda x: ''.join(x))

districts_ones.groupby('district').agg({'id':'sum'}).reset_index().sort_values(by = 'id', ascending = False)
```

Значения несколько искажены из-за потери многих данных, но тенденция ясна.
Отличились районы, приближенные к центру, где много коротких улиц.
